```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import HourLocator, DateFormatter,DayLocator,MonthLocator


from pandas import read_csv
```

```python
Burnup = read_csv("Burn Up Charts.csv")
BurnDown = read_csv("Burn Down Charts.csv")
```

```python
Burnup.head()
```

```python
sns.lineplot(data= Burnup, x = "Iteration", y = "Total Points",legend="brief", label = "Total Points")
sns.lineplot(data= Burnup, x = "Iteration", y = "MVP",legend="brief", label = "MVP")
sns.lineplot(data= Burnup, x = "Iteration", y = "Predict High",legend="brief", label = "Predict High",linestyle="dashed")
sns.lineplot(data= Burnup, x = "Iteration", y = "Predict",legend="brief", label = "Predict",linestyle="dashed")
sns.lineplot(data= Burnup, x = "Iteration", y = "Predict Low",legend="brief", label = "Predict Low",linestyle="dashed")

plt.title("Burn Up Chart: Iteration 2")
plt.grid(True)
```

```python
BurnDown.head()
```

```python
BurnDown['Date'] =  pd.to_datetime(BurnDown['Date'], format='%m/%d/%Y')
```

```python
plot=sns.lineplot(data= BurnDown, x = "Date", y = "Total Points",legend="brief", label = "Required Velocity",linestyle="dashed")
sns.lineplot(data= BurnDown, x = "Date", y = "Total Left",legend="brief", label = "Total Points Left")
sns.lineplot(data = BurnDown,x="Date",y=0, legend="brief", label = "Goal")
plt.title("Burn Down: Iteration 2")
plot.xaxis.set_major_locator(DayLocator(interval=2))
plot.xaxis.set_major_formatter(DateFormatter('%m-%d-%Y'))
plt.xticks(rotation=90)
plt.grid(visible = True)
```

```python
plot = plt.plot(BurnDown["Date"], BurnDown["Total Points"], label = "Required Velocity",linestyle="dashed")
plt.bar(BurnDown["Date"], BurnDown["Predict Low"], label = "Points Per Day", color="orange")
plt.legend(loc="upper left")
plt.show()
```
